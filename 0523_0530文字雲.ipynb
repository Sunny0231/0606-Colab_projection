```python
import nltk

from nltk.corpus import PlaintextCorpusReader

corpus_root = '/content/drive/My Drive/colab'
wordlists = PlaintextCorpusReader(corpus_root, '.*') # *: all .txt

books = wordlists.fileids()
print(books)
```

```python
import nltk

testText = wordlists.words('All Too Well.txt')
single = nltk.text.Text(testText)
print(single)
```

```python
single.count("remember")
```

```python
single.similar("innocence")
```

```python
import nltk
testText = wordlists.words('All Too Well.txt')
single = nltk.text.Text(testText)

findList = ["well","you","keep","I","it"]
single.dispersion_plot(findList)
```

```python
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:7])
    for fileid in wordlists.fileids()
    for w in wordlists.words(fileid)
    for target in [ 'positive', 'negative']
    if w.lower() == target) # w.lower().startswith(target)
cfd.plot()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


file_text = open('/content/drive/MyDrive/colab/All Too Well.txt','r').read()


wordcloud = WordCloud(
    background_color="black",
    width=2000,
    height=1000,
    margin= 10,
).generate(file_text)


plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
